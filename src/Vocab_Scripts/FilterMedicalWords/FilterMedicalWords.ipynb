```python
import argparse
from quickumls import QuickUMLS
import pandas as pd
import glob
```

```python

# parser = argparse.ArgumentParser()
# parser.add_argument('-umls_path',type=str)
# parser.add_argument('-csv_path',type=str)

umls_path = '/path/to/QuickUMLSdir'

matcher = QuickUMLS(umls_path,similarity_name='cosine') #put the default directory of UMLS here.
for csv_path in glob.glob('../GenerateCandidateWords/TokenSplitDistribution*'):
    print(f'Starting for {csv_path}.....')
    df = pd.read_csv(csv_path)
    list_consider = list()

    for idx in range(df.shape[0]):
        if idx%10000 == 0: print(f'Processed till {idx+1}...')
        
        if df.iloc[idx,1] >= 4:
            list_consider.append(1)
            continue

        flag = 0
        d = matcher.match(df.iloc[idx,0], best_match=True, ignore_syntax=False)
        if len(d) == 0:
            list_consider.append(0)
            continue


        for l in d[0]:
            if l['preferred'] == 1:
                if df.iloc[idx,1] in [2,3]:
                    if l['similarity'] >= 0.95 : 
                        flag = 1
                        break
        if flag == 1: 
            list_consider.append(1)

        else: list_consider.append(0)

    df['Consider'] = list_consider
    df.to_csv('%sWithConsiderFlag_ConsideredSemTypesWithConsiderFlag.csv'%csv_path.split('/')[-1][:-4],index=False)
```
